```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
data = []
driver = webdriver.Chrome('/Users/anis.achkar/Desktop/Uber documents/Webscarpping/chromedriver') 
for page in range(1,19):
    driver.get("https://registre-vtc.developpement-durable.gouv.fr/public/recherche-geographique/"+str(page)+"/lister")
    time.sleep(3)
    while True: 
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        table = soup.find("table", {"id": "tableListeResultat"})
        table_body = table.find('tbody')
        rows = table_body.find_all('tr')
        for row in rows:
            cols = row.find_all('td')
            cols = [ele.text.strip() for ele in cols]
            data.append([ele for ele in cols]) # Get rid of empty values
            nextButton = driver.find_element_by_xpath("//*[@id=\"tableListeResultat_next\"]")
        if "disabled" in nextButton.get_attribute("class"):
            break
        else:
            # driver.find_element_by_css_selector("li.naechste-seite").click()
            # driver.find_element("CSS_SELECTOR", value="naechste-seite") 
            # driver.find_element_by_id("tableListeResultat_next").click()
            driver.find_element_by_xpath("//*[@id=\"tableListeResultat_next\"]/a").click()
            time.sleep(1)
            page_source=driver.page_source

data=pd.DataFrame(data)
data.columns = ["Dénomination", "Nom Commercial", "Inscription", "Ville", "Région", "Actions"]
```
